```python
from sbnltk.Stemmer import stemmerOP
from sbnltk.Preprocessor import preprocessor
```

```python
def preprocess_text(text):
    p=preprocessor()
    # Removing Punctuation From Text
    text = p.punctuation_remove(text)
    print("After removing punctuation:", text)

    # Removing Stopword from text
    text = p.stopword_remove(text)
    print("After removing stopwords:", text)

    # Removing extra space
    text = p.extra_space_remove(text)
    print("After removing extra space:", text)
    return text
```

```python
def remove_suffix(text):
    st = stemmerOP()
    stemmed_text = st.stemSent(text)
    return stemmed_text
```

```python
text = "ধৃতরাষ্ট্র সঞ্জয়কে জিজ্ঞাসা করলেন: হে সঞ্জয়, আমার যোদ্ধা পুত্ররা এবং পাণ্ডুর পুত্ররা যখন ধর্মকর্মের ক্ষেত্র কুরুক্ষেত্রে সমবেত হয়েছিল তখন তারা কী করেছিল? আমাকে সেই ঘটনাগুলো বলুন।"
preprocessed_text = preprocess_text(text)
stemmed_text = remove_suffix(preprocessed_text)
print(stemmed_text)
```

```python
import pandas as pd
from collections import Counter
import nltk
```

```python
excel_file_path = "/content/Chapter-1.xlsx"
df = pd.read_excel(excel_file_path)
print(df)
```

```python
# Apply cleaning 'Shlok' column
df['PrePocesssingText'] = df['Shlok'].apply(preprocess_text)

# Stemmming Texts
df['clean-text'] = df['PrePocesssingText'].apply(remove_suffix)

print(df)
```

```python
print(df)
```

```python
# Read replacements from file
replacements_file = '/content/Stop_Word_ReplaceGITA.txt'
replacements = {}
with open(replacements_file, 'r', encoding='utf-8') as file:
    for line in file:
        old_word, new_word = line.strip().split(':')
        replacements[old_word] = new_word

def apply_replacements(text):
    for old_word, new_word in replacements.items():
        text = text.replace(old_word, new_word)
    return text

# Apply the function to the 'clean-text' column
df['clean-text'] = df['clean-text'].apply(apply_replacements)
```

```python
def read_words_to_remove(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        words_to_remove = [line.strip() for line in file]
    return words_to_remove

def apply_removals(text, words_to_remove):
    for word in words_to_remove:
        text = text.replace(word, '')
    return text

# Read words to remove from file
file_path = '/content/stopwordsGITA.txt'
words_to_remove = read_words_to_remove(file_path)

# Apply the function to the 'clean-text' column
df['clean-text'] = df['clean-text'].apply(lambda text: apply_removals(text, words_to_remove))
```

```python
# Function to count words from text
def count_words(text):
    tokens = nltk.word_tokenize(text)
    return Counter(tokens)

# Concatenate all text from the 'TEXT' column into a single string
all_text = ' '.join(df['clean-text'].astype(str))

# Count words and their occurrences
word_count = count_words(all_text)
# Sort word count from high to low
sorted_word_count = dict(sorted(word_count.items(), key=lambda item: item[1], reverse=True))

# Display word count results
print("Word\t\tCount")
for word, count in sorted_word_count.items():
    print(f"{word}\t\t{count}")


output_file_path = "/content/Output/word_count_results.txt"
with open(output_file_path, 'w', encoding='utf-8') as file:
    file.write("Word\t\tCount\n")
    for word, count in sorted_word_count.items():
        file.write(f"{word}\t\t{count}\n")

print("Word count results saved to:", output_file_path)
```

```python
# Print total unique word count
total_unique_words = len(sorted_word_count)
print(f"Total unique words: {total_unique_words}")

# Print top 10 words and their counts
print("\nTop 10 words:")
print("Word\t\tCount")
top_10_words = list(sorted_word_count.items())[:10]
for word, count in top_10_words:
    print(f"{word}\t\t{count}")
```

```python
# Save the results to a new Excel file
output_excel_path = "/content/Output/Cleaned_Chapter-1.xlsx"
df.to_excel(output_excel_path, index=False)

print("Cleaned data saved to", output_excel_path)
```
